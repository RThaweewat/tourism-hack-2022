```python
def get_heatmap(df1:pd.DataFrame, df2:pd.DataFrame, filter_by:str, filter_by_2:str, count:bool, min:float, max:float):
    df_copied = df1.copy()
    df_copied.loc[:, 'LATITUDE'] = ((df_copied.loc[:, 'hotel_latitude'] * 1000).apply(np.floor)) / 1000
    df_copied.loc[:, 'LONGITUDE'] = ((df_copied.loc[:, 'hotel_longitude'] * 1000).apply(np.floor)) / 1000
    df_copied.loc[:, 'LatLonRange'] = df_copied.loc[:, 'hotel_latitude'].map(str) + '-' + df_copied.loc[:, 'hotel_longitude'].map(str)

    df_copied2 = df2.copy()
    df_copied2.loc[:, 'LATITUDE'] = ((df_copied2.loc[:, 'latitude'] * 100).apply(np.floor)) / 100
    df_copied2.loc[:, 'LONGITUDE'] = ((df_copied2.loc[:, 'longitude'] * 100).apply(np.floor)) / 100
    df_copied2.loc[:, 'LatLonRange'] = df_copied2.loc[:, 'latitude'].map(str) + '-' + df_copied2.loc[:, 'longitude'].map(str)

    df_grouped = df_copied.groupby(['LatLonRange', 'LATITUDE', 'LONGITUDE'])
    df_grouped2 = df_copied2.groupby(['LatLonRange', 'LATITUDE', 'LONGITUDE'])

    if count == True:
        filter_count = df_grouped[filter_by].agg(['count']).reset_index()
        filter_count2 = df_grouped2[filter_by_2].agg(['count']).reset_index()
        df_folium = pd.DataFrame({'Lat':filter_count['LATITUDE'],'Long':filter_count['LONGITUDE'],'Count':filter_count['count']})
        df_folium2 = pd.DataFrame({'Lat':filter_count2['LATITUDE'],'Long':filter_count2['LONGITUDE'],'Count':filter_count2['count']})
        df_folium['weight'] = df_folium['Count'] / df_folium['Count'].abs().max()
        df_folium2['weight'] = df_folium2['Count'] / df_folium2['Count'].abs().max()
    else:
        filter_avgSize = df_grouped[filter_by].agg(['mean']).reset_index()
        filter_avgSize2 = df_grouped2[filter_by_2].agg(['mean']).reset_index()
        df_folium = pd.DataFrame({'Lat':filter_avgSize['LATITUDE'],'Long':filter_avgSize['LONGITUDE'],'Count':filter_avgSize['mean']})
        df_folium2 = pd.DataFrame({'Lat':filter_avgSize2['LATITUDE'],'Long':filter_avgSize2['LONGITUDE'],'Count':filter_avgSize2['mean']})
        df_folium['weight'] = df_folium['Count'] / df_folium['Count'].abs().max()
        df_folium2['weight'] = df_folium2['Count'] / df_folium2['Count'].abs().max()


    def generateBaseMap(loc, zoom=11, tiles="Stamen Terrain", arrr = None,crs='ESPG2263'):
        return folium.Map(location=loc,
                          control_scale=True,
                          zoom_start=zoom,
                          tiles=tiles,
                          attr=arrr)

    base_map = generateBaseMap([7.95, 98.33] )
    map_values1 = df_folium[['Lat','Long','weight']]
    map_values2 = df_folium2[['Lat','Long','weight']]
    data = map_values1.values.tolist()
    data2 = map_values2.values.tolist()
    hm = HeatMap(data,gradient={0.1: 'white',  0.5: 'yellow', 1: 'red'},
                 min_opacity=min,
                 max_opacity=max,
                 blur=15,
                 radius=30,
                 use_local_extrema=True) #.add_to(base_map)

    hm2 = HeatMap(data2,gradient={0.1: 'white', 1: 'grey'},
                 min_opacity=min,
                 max_opacity=max,
                 blur=20,
                 radius=30,
                 use_local_extrema=False)

    for index, location_info in df_copied2.dropna().iterrows():
	    folium.Marker([location_info["latitude"], location_info["longitude"]], popup=location_info["title"]).add_to(base_map)

    return base_map.add_child(hm).add_child(hm2)
```
